```ruby
full_df = Rover.read_csv('../Data_sources/Data/In_sample/ETHUSD.csv')
full_df['candle_size'] = full_df['High'] - full_df['Low']
full_df['candle_size_ma'] = 0.0
full_df['direction'] = ''

(24..full_df.size-1).each do |i|
  full_df['candle_size_ma'][i] = full_df['candle_size'][i-24..i].mean
  full_df['direction'][i] = full_df['Close'][i] > full_df['Open'][i] ? 'bull' : 'bear'
end
df = full_df[24..-1]
```

```ruby
def format_time(time)
  DateTime.strptime(time, "%Y-%m-%d %H:%M:%S")
end
```

```ruby
def daily_range(df)
  _daily_range = []
  df['Datetime'].each_with_index do |item, index|
    if format_time(item).day != format_time(df['Datetime'][index-1]).day
      begin
        _daily_range << df['High'][index..index+24].max - df['Low'][index..index+24].min
      rescue
        _daily_range << df['High'][index..-1].max - df['Low'][index..-1].min
      end
    end
  end
  r = Rover::DataFrame.new({ daily_range: _daily_range})
  r[:daily_range][-10..-1].mean
end
daily_range(df)
```
