```ruby
require 'httpx'
require 'rover-df'
url = 'https://api.binance.com/api/v3/klines' +'?symbol=' + 'BTCUSDT' + '&interval=' + '1h'
page = HTTPX.get(url)
```

```ruby
columns = [:datetime, :open, :high, :low, :close, :volume]
data = page.json.map do |i|
  e = {}
  columns.each do |name|
    e[name] = i.shift
  end
  e
end
df = Rover::DataFrame.new data
# df[:datetime] = df[:datetime].map {|i| Time.at(i/1000)}
[:open, :high, :low, :close, :volume].each { |i| df[i] = df[i].map(&:to_f) }
df
df.plot(:datetime, :close, type: "line")
df[10]
Vega.lite
  .data(df)
  .mark(type: "line")
  .encoding(
    x: {field: :datetime, type: 'temporal'},
    y: {field: :close, type: 'quantitative', scale: { domain: [18000,25000] }}
  )
```
